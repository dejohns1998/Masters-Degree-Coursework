That said, it seems reasonable that there would be a small comparable number of records with income or populations above 100,000, as examples. 

>__Regarding Data Quality__: Some variables contain misleading data entries which may generate skewed results during the analysis phase. The *marital_status* variable has the same issue. 

## D7: Impact of Limitations

>__Impact of Null Values__: Replacing null values with measures of central tendency for each variable could potentially lead to misleading results when answering our research question, as it could inaccurately ascribe correlation with certain results. For instance, there were null values in *soft_drink_consumption*, *overweight*, and *anxiety* which are now replaced with the median for each of those respective categories. This may lead to these variables being inaccurately relationed to readmission during the data analysis phase. That said, these values appear to be missing at random (MAR) and I don't expect it to have a significant effect on the overall data distribution.

>__Impact of Outliers__: It is possible that some of the outliers identified are in fact misinput entries, in which case, this could again lead to misleading results when answering our research question, as it could inaccurately ascribe correlation with certain results. ## E1: Total Number of Principal Components

```python
#Subset the data to include only numeric data
df_pca = df[["latitude", "longitude", "population", "children_in_household", "age", "sponsor_income", "vitd_lvl", "pcm_visits", "hosp_meals_eaten", "vitd_supp", "initial_stay", "est_daily_charge", "est_additional_charges"]]
# Normalize data
df_pca_normalized = (df_pca-df_pca.mean())/df_pca.std()
#How many components to extract
pca = PCA(n_components=df_pca.shape[1])
#Create new dataset converting variables to compoenents
pca.fit(df_pca_normalized)
pca_printable = pd.DataFrame(pca.transform(df_pca_normalized), columns=["PC1", "PC2", "PC3", "PC4", "PC5", 
                                                                 "PC6", "PC7", "PC8", "PC9", "PC10", "PC11", "PC12", "PC13"])
#Generate loadings for the components
pca_loadings = pd.DataFrame(pca.components_.T, columns = ["PC1", "PC2", "PC3", "PC4", "PC5", 
                                                          "PC6", "PC7", "PC8", "PC9", "PC10", "PC11", "PC12", "PC13"],
                            index=df_pca_normalized.columns)
pca_loadings
```

## E2: Justification of Principal Components and Scree Plot

```python
cov_matrix = np.dot(df_pca_normalized.T, df_pca_normalized) / df_pca.shape[0]
eigenvalues = [np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)) for eigenvector in pca.components_]

plt.plot(eigenvalues)
plt.xlabel('number of components')
plt.ylabel('eigenvalue')
plt.show() 
```

__Reduced Number of Principal Components__: Using the scree plot, we can see there are six principal components.
<br>
<br>
__Justification of Reduced Principal Components__: Eigenvalues demonstrate the total variance that can be explained by a principal component - the higher the eigenvalue the more the component captures a significant part of the dataset's structure. Meanwhile, a lower eigenvalue adds less valuable information as it captures a smaller proportion of the total variability ("Principal components (pca) and exploratory factor analysis (efa) with spss," n.d.). Here, we can see six components with eigenvalues above '1'. PC7 - PC13 have eigenvalues less than 1 and can be discarded, thus reducing the dimensionality of the data while retaining the most critical information.

## E3: Benefits of Principal Component Analysis

We are dealing with a rather large dataset with more than 50 variables and 10,000 records. Conducting Principal Component Analysis highlights the most significant features of our data, facilitating future data visualization and analysis. 

PCA operates by transforming the original variables of the dataset into a new set of 'principal components.' Each of these is a linear combination of the original variables, capturing the maximum possible variance ordered from largest to smallest variance. The eigenvalues represent the level of variance, the higher, the more variance, the more significance (Connelly, 2011).

PCA reduces the dimensionality of our data, thus helping us to avoid issues further on in the data cycle by reducing computational load and avoiding overfitting (when a model includes too much noise and operates better on training data than on new, potentially "real-world" data). This simplification helps in future data analysis and improve machine learning performance. By revealing the underlying structure of the data, PCA can identify statistically significant relationships which will help us to better identify the most influential factors behind patient readmission (Seo, 2023).
