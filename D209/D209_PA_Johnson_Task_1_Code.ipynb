```python
#Establishing standardized pythonic naming convention for variables
new_columns = ["customer_id", "interaction", "uid", "city", "state", "county", 
                    "zip", "latitude", "longitude", "population", "area", "timezone", "job", "children_in_household", 
                    "age", "sponsor_income", "marital_status", "gender", "readmis_within_month", 
                    "vitd_lvl", "pcm_visits", "hosp_meals_eaten", "vitd_supp", "soft_drink_consumption", "initial_admin", 
                    "high_blood_pressure", "stroke", "complication_risk", "overweight", "arthritis", "diabetes", "hyperlipidemia", 
                    "back_pain", "anxiety", "allergic_rhinitis", "reflux_esophagitis", "asthma", "services", 
                    "initial_stay", "est_daily_charge", "est_additional_charges", "likert_timely_admission", "likert_timely_treatment", 
                    "likert_timely_visit", "likert_reliablity", "likert_options", "likert_treatment_hrs", "likert_courteous_staff", 
                    "likert_dr_active_listening"]
df.columns = new_columns
print(df.columns)
```

```python
## Adjusting datatypes as necessary 

#Transforming state to categorical 
df["state"] = df["state"].astype("category")

#Transforming zip to string
df['zip'] = df['zip'].astype("str").str.zfill(5)

#Transforming area to categorical
df["area"] = df["area"].astype("category")

#Transforming timezone to categorical
df["timezone"] = df["timezone"].astype("category")

#Transforming children_in_household to int64
df["children_in_household"] = df["children_in_household"].astype("Int64")

#Transforming marital_status to categorical 
df["marital_status"] = df["marital_status"].astype("category")

#Transforming gender to categorical 
df["gender"] = df["gender"].astype("category")

#Transforming soft_drink_consumption to boolean
df["soft_drink_consumption"] = df["soft_drink_consumption"].astype("bool")

#Transforming initial_admin to categorical
df["initial_admin"] = df["initial_admin"].astype("category")

#Transforming complication_risk to categorical
df["complication_risk"] = df["complication_risk"].astype("category")

#Transforming services to categorical
df["services"] = df["services"].astype("category")

#Transforming initial_stay to int64
df["initial_stay"] = df.initial_stay.round(2)
df["initial_stay"] = df["initial_stay"].astype("int64")

#Transforming est_daily_charge to int64
df["est_daily_charge"] = df.est_daily_charge.round(2)
df["est_daily_charge"] = df["est_daily_charge"].astype("int64")

#Transforming est_additional_charges to int64
df["est_additional_charges"] = df.est_additional_charges.round(2)
df["est_additional_charges"] = df["est_additional_charges"].astype("int64")

#Transforming likert_timely_admission to categorical
df["likert_timely_admission"] = df["likert_timely_admission"].astype("category")

#Transforming likert_timely_treatment to categorical
df["likert_timely_treatment"] = df["likert_timely_treatment"].astype("category")

#Transforming likert_timely_visit to categorical
df["likert_timely_visit"] = df["likert_timely_visit"].astype("category")

#Transforming likert_reliablity to categorical
df["likert_reliablity"] = df["likert_reliablity"].astype("category")

#Transforming likert_options to categorical
df["likert_options"] = df["likert_options"].astype("category")

#Transforming likert_treatment_hrs to categorical
df["likert_treatment_hrs"] = df["likert_treatment_hrs"].astype("category")

#Transforming likert_courteous_staff to categorical 
df["likert_courteous_staff"] = df["likert_courteous_staff"].astype("category")

#Transforming likert_dr_active_listening to categorical
df["likert_dr_active_listening"] = df["likert_dr_active_listening"].astype("category")
```

```python
## Transformation of data for classification 
from sklearn.preprocessing import LabelEncoder

# Convert categorical variables to numerical values
le = LabelEncoder()
categorical_features = ['high_blood_pressure', 'stroke', 'overweight', 'arthritis', 'diabetes', 'hyperlipidemia', 'back_pain', 'anxiety', 'allergic_rhinitis', 'reflux_esophagitis', 'asthma', 'readmis_within_month']
for feature in categorical_features:
    df[feature] = le.fit_transform(df[feature])

# Reduce dataset
nb_df = df[["high_blood_pressure", "stroke", "overweight", "arthritis", "diabetes", "hyperlipidemia", "back_pain", "anxiety", "allergic_rhinitis", "reflux_esophagitis", "asthma", "readmis_within_month"]]
 
# Dummies for complication_risk
comp_risk = pd.get_dummies(data=df["complication_risk"], drop_first=True, dtype=int)
nb_df.insert(4, "comp_risk_medium", comp_risk.Medium)
nb_df.insert(4, "comp_risk_low", comp_risk.Low)

#Check dataset output
nb_df
```

## D1: Training and Test Data

```python
#Creation of synethetic dataset for training model
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt


X, y = make_classification(n_samples=100, n_features=2,
                           n_redundant=0, n_clusters_per_class=1,
                           random_state=42)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('Training Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Training the model using the synthetic dataset
from sklearn.model_selection import train_test_split

# Extract features and labels
X = nb_df.drop('readmis_within_month', axis=1)
y = nb_df['readmis_within_month']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Save each of the four data sets to CSV
pd.DataFrame(X_train).to_csv('task1_Xtrain.csv', index=False)
pd.DataFrame(X_test).to_csv('task1_Xtest.csv', index=False)
pd.DataFrame(y_train).to_csv('task1_ytrain.csv', index=False)
pd.DataFrame(y_test).to_csv('task1_ytest.csv', index=False)
```

```python
#Checking output of test dataset to ensure proper split
print(X_test[:3])
```

## D2: Analysis Technique

```python
print(pd.Series(y_train).value_counts(normalize=True))
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, f1_score

# Initialize the Gaussian Naive Bayes classifier
bnb = BernoulliNB()

# Train the model
bnb.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = bnb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

#Calculate F1 Score
f1 = f1_score(y_pred, y_test, average="weighted")
print("F1 Score:", f1)
```

```python
#Calculating feature importance
import numpy as np 
prob_pos = pd.Series(y_test).value_counts(normalize=True)[0]
prob_neg = pd.Series(y_test).value_counts(normalize=True)[1]

#Creating a dataframe to store the words
df_nbf = pd.DataFrame()
df_nbf.index = X.columns 

# Convert log probabilities to probabilities. 
df_nbf['pos'] = np.e**(bnb.feature_log_prob_[0, :])
df_nbf['neg'] = np.e**(bnb.feature_log_prob_[1, :])

#Calculating odds ratio
df_nbf['odds_positive'] = (df_nbf['pos']/df_nbf['neg'])*(prob_pos /prob_neg)
df_nbf['odds_negative'] = (df_nbf['neg']/df_nbf['pos'])*(prob_neg/prob_pos)

#Converting to a percentage to better show influence
df_nbf['percentage_increase'] = (df_nbf['odds_positive'] - 1) * 100
print(df_nbf)
```

## D3: Classification Analysis Code

```python
##Calculate accuracy at different thresholds

from sklearn.metrics import confusion_matrix, roc_auc_score

# Get probability predictions
y_proba = bnb.predict_proba(X_test)[:, 1]

# Threshold of 0.37
threshold = 0.37
y_pred_adjusted_1 = (y_proba > threshold).astype(int) 

# Threshold of 0.5
threshold = 0.5
y_pred_adjusted_3 = (y_proba > threshold).astype(int)

# Calculate the accuracy with adjusted threshold
accuracy = accuracy_score(y_test, y_pred_adjusted_1)
print(f'Accuracy with threshold of 0.37: {accuracy}')

accuracy = accuracy_score(y_test, y_pred_adjusted_3)
print(f'Accuracy with threshold of 0.5: {accuracy}')
```

```python
# Evaluate with threshold of 0.37
print("Confusion Matrix for 0.37 Threshold:")
print(confusion_matrix(y_test, y_pred_adjusted_1))

# Evaluate with threshold of 0.5
print("Confusion Matrix for 0.5 Threshold:")
print(confusion_matrix(y_test, y_pred_adjusted_3))
```

```python
# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, y_proba)
print(roc_auc)
```
