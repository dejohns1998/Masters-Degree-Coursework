```python
#Establishing standardized pythonic naming convention for variables
new_columns = ["customer_id", "interaction", "uid", "city", "state", "county", 
                    "zip", "latitude", "longitude", "population", "area", "timezone", "job", "children_in_household", 
                    "age", "sponsor_income", "marital_status", "gender", "readmis_within_month", 
                    "vitd_lvl", "pcm_visits", "hosp_meals_eaten", "vitd_supp", "soft_drink_consumption", "initial_admin", 
                    "high_blood_pressure", "stroke", "complication_risk", "overweight", "arthritis", "diabetes", "hyperlipidemia", 
                    "back_pain", "anxiety", "allergic_rhinitis", "reflux_esophagitis", "asthma", "services", 
                    "initial_stay", "est_daily_charge", "est_additional_charges", "likert_timely_admission", "likert_timely_treatment", 
                    "likert_timely_visit", "likert_reliablity", "likert_options", "likert_treatment_hrs", "likert_courteous_staff", 
                    "likert_dr_active_listening"]
df.columns = new_columns
print(df.columns)
```

```python
## Adjusting datatypes as necessary 

#Transforming state to categorical 
df["state"] = df["state"].astype("category")

#Transforming zip to string
df['zip'] = df['zip'].astype("str").str.zfill(5)

#Transforming area to categorical
df["area"] = df["area"].astype("category")

#Transforming timezone to categorical
df["timezone"] = df["timezone"].astype("category")

#Transforming children_in_household to int64
df["children_in_household"] = df["children_in_household"].astype("Int64")

#Transforming marital_status to categorical 
df["marital_status"] = df["marital_status"].astype("category")

#Transforming gender to categorical 
df["gender"] = df["gender"].astype("category")

#Transforming soft_drink_consumption to boolean
df["soft_drink_consumption"] = df["soft_drink_consumption"].astype("bool")

#Transforming initial_admin to categorical
df["initial_admin"] = df["initial_admin"].astype("category")

#Transforming complication_risk to categorical
df["complication_risk"] = df["complication_risk"].astype("category")

#Transforming services to categorical
df["services"] = df["services"].astype("category")

#Transforming initial_stay to int64
df["initial_stay"] = df.initial_stay.round(2)
df["initial_stay"] = df["initial_stay"].astype("int64")

#Transforming est_daily_charge to int64
df["est_daily_charge"] = df.est_daily_charge.round(2)
df["est_daily_charge"] = df["est_daily_charge"].astype("int64")

#Transforming est_additional_charges to int64
df["est_additional_charges"] = df.est_additional_charges.round(2)
df["est_additional_charges"] = df["est_additional_charges"].astype("int64")

#Transforming likert_timely_admission to categorical
df["likert_timely_admission"] = df["likert_timely_admission"].astype("category")

#Transforming likert_timely_treatment to categorical
df["likert_timely_treatment"] = df["likert_timely_treatment"].astype("category")

#Transforming likert_timely_visit to categorical
df["likert_timely_visit"] = df["likert_timely_visit"].astype("category")

#Transforming likert_reliablity to categorical
df["likert_reliablity"] = df["likert_reliablity"].astype("category")

#Transforming likert_options to categorical
df["likert_options"] = df["likert_options"].astype("category")

#Transforming likert_treatment_hrs to categorical
df["likert_treatment_hrs"] = df["likert_treatment_hrs"].astype("category")

#Transforming likert_courteous_staff to categorical 
df["likert_courteous_staff"] = df["likert_courteous_staff"].astype("category")

#Transforming likert_dr_active_listening to categorical
df["likert_dr_active_listening"] = df["likert_dr_active_listening"].astype("category")
```

```python
## Transformation of data for classification 
from sklearn.preprocessing import LabelEncoder

# Convert categorical variables to numerical values
le = LabelEncoder()
features = ['age', 'high_blood_pressure', 'stroke', 'overweight', 'arthritis', 'diabetes', 'hyperlipidemia', 'back_pain', 'anxiety', 'allergic_rhinitis', 'reflux_esophagitis', 'asthma', 'readmis_within_month']
for feature in features:
    df[feature] = le.fit_transform(df[feature])

# Reduce dataset
ct_df = df[["age", "high_blood_pressure", "stroke", "overweight", "arthritis", "diabetes", "hyperlipidemia", "back_pain", "anxiety", "allergic_rhinitis", "reflux_esophagitis", "asthma", "readmis_within_month"]]
 
# Dummies for complication_risk
comp_risk = pd.get_dummies(data=df["complication_risk"], dtype=int)
ct_df.insert(4, "comp_risk_medium", comp_risk.Medium)
ct_df.insert(4, "comp_risk_low", comp_risk.Low)

# Dummies for gender
gender = pd.get_dummies(data=df["gender"], dtype=int)
ct_df.insert(5, "gender_male", gender.Male)
ct_df.insert(5, "gender_female", gender.Female)

#Check dataset output
ct_df
```

## D1: Training and Test Data

```python
# Importing the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree, metrics
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, precision_recall_curve, mean_squared_error, roc_auc_score 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Separate the X variables (explanatory) from the Y variable (response) into different dataframes
X = ct_df.drop(["readmis_within_month"], axis=1).copy()
y = ct_df["readmis_within_month"].copy()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Save each of the four data sets to CSV
pd.DataFrame(X_train).to_csv('task2_Xtrain.csv', index=False)
pd.DataFrame(X_test).to_csv('task2_Xtest.csv', index=False)
pd.DataFrame(y_train).to_csv('task2_ytrain.csv', index=False)
pd.DataFrame(y_test).to_csv('task2_ytest.csv', index=False)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train the model
clf.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = clf.predict(X_test)
```

```python
# Create a decision tree with weights for the readmission class
clf_weighted = DecisionTreeClassifier(
    class_weight={0: 1, 1: 10},  
    min_samples_leaf=20,         
    max_depth=3                  
)
clf_weighted.fit(X_train, y_train)  

# Visualize this weighted tree
plt.figure(figsize=(15, 8))
tree.plot_tree(clf_weighted, 
              filled=True, 
              feature_names=list(X.columns),
              class_names=['Not Readmitted', 'Readmitted'])
plt.savefig('weighted_decision_tree.png')
plt.show()
```

## D3: Classification Analysis Code

```python
# Print classification report
print(f'Classification Report: \n{classification_report(y_test, y_pred)}')
```

```python
# Print model accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Print mean squared error to satisfy rubric
print("MSE:",mean_squared_error(y_test,y_pred))
```

```python
import numpy as np

# Compute Binary Cross-Entropy using NumPy only
def binary_cross_entropy(y_test, y_pred):
    epsilon = 1e-15  # To avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    loss = -(y_test * np.log(y_pred) + (1 - y_test) * np.log(1 - y_pred)).mean()
    return loss

print("Binary Cross-Entropy:", binary_cross_entropy(y_test, y_pred))
```

```python
# Print confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
```

```python
# Plot Precision-Recall Curve
y_pred_proba = clf.predict_proba(X_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(recall, precision, label='Decision Tree Classification', color = 'firebrick')
ax.set_title('Precision-Recall Curve')
ax.set_xlabel('Recall')
ax.set_ylabel('Precision')
plt.box(False)
ax.legend();
```

```python
# Plot AUC/ROC curve
y_pred_proba = clf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_test,  y_pred_proba)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(fpr, tpr, label='Decision Tree Classification', color = 'firebrick')
ax.set_title('ROC Curve')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
plt.box(False)
ax.legend();
```
